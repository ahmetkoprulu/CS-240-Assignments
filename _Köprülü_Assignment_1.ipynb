```python
# First Import these files

from __future__ import print_function, division
import numpy as np
import nsfg
import thinkplot
import thinkstats2
```

```python
resp = nsfg.ReadFemResp()
print(resp.head(20))
resp.tail(30)
```

```python
ages = resp.agescrn
sorted = ages.value_counts().sort_index()

print(sorted)
print("Biggest age is " + str(sorted.keys()[0]) + ", Smallest age is " + str(sorted.keys()[len(sorted)-1]))
```

```python
agesLessThan25 = resp[resp.agescrn <= 25]
agesLessThan25.numpregs.mean()
```

```python
marriage = resp.fmarno
#print(marriage)
hist = thinkstats2.Hist(marriage)
thinkplot.Hist(hist)
```

```python
nevermarried = resp[resp.numpregs == 0]
married = resp[resp.numpregs > 0]

nevermarried_hist = thinkstats2.Hist(nevermarried.totincr, label='Never Married')
married_hist = thinkstats2.Hist(married.totincr, label='Married')

width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(married_hist, align='right', width=width)
thinkplot.Hist(nevermarried_hist, align='left', width=width)
thinkplot.Config(xlabel='Incomes', ylabel='Frequency')
```

```python
nevermarried_min = nevermarried.totincr.min()
nevermarried_max = nevermarried.totincr.max()
nevermarried_mean = nevermarried.totincr.mean()
nevermarried_std = nevermarried.totincr.std()
nevermarried_var = nevermarried.totincr.var()

married_min = married.totincr.min()
married_max = married.totincr.max()
married_mean = married.totincr.mean()
married_std = married.totincr.std()
married_var = married.totincr.var()

print('Minimum Never married total income is ' + str(nevermarried_min) + ' | Maximum Never married total income is  ' + str(nevermarried_max))
print('Mean of Never married total income is ' + str(nevermarried_mean) + ' | Varience of Never married total income is  ' + str(nevermarried_var))
print('Standart Derivation of Never married total income is  ' + str(nevermarried_std))
print("")
print('Minimum married total income is ' + str(married_min) + ' | Maximum married total income is  ' + str(married_max))
print('Mean of married total income is ' + str(married_mean) + ' | Varience of married total income is  ' + str(married_var))
print('Standart Derivation of married total income is  ' + str(married_std))

```
