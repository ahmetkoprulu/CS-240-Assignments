```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import analytic

import thinkstats2
import thinkplot
```

# Question 1

```python
resp = nsfg.ReadFemResp()

university = resp[resp.hieduc >= 11]
school = resp[resp.hieduc <= 10]
```

```python
university_max = university.agemomb1.max()
university_mean = university.agemomb1.mean()
university_std = university.agemomb1.std()
school_max = school.agemomb1.max()
school_mean = school.agemomb1.mean()

print ('Avarage age of mothers who had first for degree is', university_mean)
print ('Maximum age of mothers who had first for degree is', university_max)
print ('Avarage age of mothers who had first for not degree is', school_mean)
print ('Maximum age of mothers who had first for not degree is', school_max)
print ('Standard deviasion age of mothers who had first is', university_std)
```

# Question 2

```python
cdf_university = thinkstats2.Cdf(university.agemomb1)
cdf_school = thinkstats2.Cdf(school.agemomb1)
thinkplot.Cdf(cdf_university, label='Degree')

thinkplot.preplot(1)
thinkplot.Cdf(cdf_school, label='Not Degree')

thinkplot.Config(title='Exponential CDF', xlabel='ages of mothers at first birth', ylabel='CDF', 
                 loc='lower right')
```

```python
pmf_university = thinkstats2.Pmf(university.agemomb1)
pmf_school = thinkstats2.Pmf(school.agemomb1)
thinkplot.Pmf(pmf_university, label='Degree')

thinkplot.preplot(1)
thinkplot.Pmf(pmf_school, label='Not Degree')

thinkplot.Config(title='PMF', xlabel='ages of mothers at first birth', ylabel='PMF', 
                 loc='upper right')
```

## Question 4

```python
cdf_university = thinkstats2.Cdf(university.agemomb1)
cdf_school = thinkstats2.Cdf(school.agemomb1)

thinkplot.PrePlot(2)
thinkplot.Cdfs([cdf_university, cdf_school])

thinkplot.Config(title='Exponential CDF', xlabel='ages of mothers at first birth', ylabel='CDF', 
                 loc='lower right')
```

```python
percentilesc = cdf_school.PercentileRank(1)
percentileuni = cdf_university.PercentileRank(1)
probsc = cdf_school.Prob(1)
probuni = cdf_university.Prob(1)

print ('Probability of mothers who had first at less than 18 years old for no degree is', probsc)
print ('Percentage of mothers who are 18 or older in for no degree is', 100-percentilesc)

print ('Probability of mothers who had first at less than 18 years old for degree is', probuni)
print ('Percentage of mothers who are 18 or older in for degree is', 100-percentileuni)
```
